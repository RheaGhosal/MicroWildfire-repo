## 3) Load bootstrap CI results and show a quick chart

```python
with open('out/bootstrap_ci.json') as f:
    ci = json.load(f)
ci
```

```python
vals = ci['auroc']  # [mean, lo, hi]
mean_auc, lo, hi = vals
err_lo = mean_auc - lo
err_hi = hi - mean_auc
plt.figure()
plt.errorbar([0],[mean_auc], yerr=[[err_lo],[err_hi]])
plt.title('AUROC (mean ± 95% CI)')
plt.ylabel('AUROC')
plt.xticks([])
plt.show()
```
